#Data preprocessing

##Detecting outliers

```python
def outlier_detect(df):
    for i in df.describe().columns:
        Q1=df.describe().at['25%',i]
        Q3=df.describe().at['75%',i]
        IQR=Q3 - Q1
        LB=Q1 - 1.5 * IQR
        UB=Q3 + 1.5 * IQR
        x=np.array(df[i])
        p=[]
        for j in x:
            if j < LB or j>UB:
               p.append(j)
        print('\n Outliers for Column : ', i, ' Outliers count ', len(p))
        print(p)
```

```python
,numeric_columns     = ['age', 'bmi', 'children', 'charges']
x=df[numeric_columns]
outlier_detect(x)
```

##Encoding categorical variables

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['sex'] = df['sex'].map(lambda s :1  if s == 'female' else 0)
df['smoker'] = df['smoker'].map(lambda s :1  if s == 'yes' else 0)
df['region']= label_encoder.fit_transform(df['region'])
df.head()
```

```python
#splitting the data into dependent and independent variables
X = df.iloc[:,:-1]
y = df['insuranceclaim']
```

```python
X.head()
```

```python
y.head()
```

##Feature selection using Extra Tree Classifier


```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
#Calculating feature importance
print(model.feature_importances_)
```

```python
#Ranking features according to importance
ranked_features = pd.Series(model.feature_importances_, index=X.columns)
ranked_features.nlargest(len(X.columns)).plot(kind='bar')
plt.show()
```

```python
#Splitting into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.head())
print(X_test.head())
print(y_train.head())
print(y_test.head())
```

#Classification

##Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
conf_clf = confusion_matrix(y_test,y_pred)
conf_clf
```

```python
#Classification Report
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print(clf_report)
```

```python
#Predicting decision tree accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Visualizing decision tree
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus
feature_cols = ['age',	'sex',	'bmi',	'children',	'smoker',	'region',	'charges']
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('insurance_dt.png')
Image(graph.create_png())
```
